# Basic Line Chart

```python
import matplotlib.pyplot as plt

# Arrays for X and Y coordinates
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create line chart
plt.plot(x, y, color='blue', linestyle='--', linewidth=2)

# Customize the appearance
plt.title('Basic Line Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

# Show the plot
plt.show()
```

# Chart Customisation

```python
# Create plot
plt.plot(x, y, color='green', marker='o')

# Customize the appearance
plt.title('Customized Line Chart', fontdict={'fontsize': 16, 'fontweight': 'bold', 'fontstyle': 'italic'})
plt.xlabel('X-axis', fontdict={'fontsize': 12, 'fontweight': 'normal', 'fontstyle': 'italic'})
plt.ylabel('Y-axis', fontdict={'fontsize': 12, 'fontweight': 'normal', 'fontstyle': 'italic'})
plt.grid(True)

# Show plot
plt.show()
```

# Advanced Chart Features

```python
# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [3, 6, 9, 12, 15]

# Create plot with three different data series
plt.plot(x, y1, color='blue', linestyle='-', marker='o', label='Series 1')
plt.plot(x, y2, color='green', linestyle='--', marker='s', label='Series 2')
plt.plot(x, y3, color='red', linestyle=':', marker='^', label='Series 3')

# Customize the appearance
plt.title('Multiple Data Series')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

# Add legend with custom location and font size
plt.legend(loc='upper left', fontsize='large')

# Show plot
plt.show()
```

# Resizing and Saving Charts

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create plot
plt.plot(x, y, color='blue', marker='o')
plt.title('Resized and Saved Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

# Resize plot to specific size
plt.figure(figsize=(8, 6))  # Width: 8 inches, Height: 6 inches

# Save plot to PNG format with high resolution
plt.savefig('resized_plot.png', dpi=300)  # DPI: 300 (high resolution)

# Show plot
plt.show()
```

# Bar Chart Creation

```python
# Sample data
products = ['Product A', 'Product B', 'Product C']
average_sales = [5000, 7000, 6000]  # Average monthly sales for each product

# Create bar chart
plt.bar(products, average_sales, color=['blue', 'green', 'red'], hatch=['/', '+', '-'])

# Customize appearance
plt.title('Average Monthly Sales of Products')
plt.xlabel('Products')
plt.ylabel('Average Sales')
plt.grid(axis='y')

# Show plot
plt.show()
```

# Multiple Line Plots

```python
# Sample data
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]  # Data for line 1
y2 = [100, 200, 300, 400, 500]  # Data for line 2

# Create figure and axes
fig, ax1 = plt.subplots()

# Plot line 1 with left Y-axis
color = 'tab:blue'
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis 1', color=color)
ax1.plot(x, y1, color=color, marker= 'o')
ax1.tick_params(axis='y', labelcolor=color)

# Create another Y-axis for line 2
ax2 = ax1.twinx()

# Plot line 2 with right Y-axis
color = 'tab:red'
ax2.set_ylabel('Y-axis 2', color=color)
ax2.plot(x, y2, color=color, marker='*')
ax2.tick_params(axis='y', labelcolor=color)

# Customize appearance
plt.title('Multiple Line Plots with Different Y-axes')

# Show plot
plt.show()
```

# Pie Chart Customisation

```python
# Sample data
brands = ['Brand A', 'Brand B', 'Brand C', 'Brand D', 'Brand E']
market_share = [30, 20, 15, 10, 25]  # Market share percentage for each brand

# Explode the largest segment (Brand A)
explode = (0.2, 0, 0, 0, 0)

# Create pie chart
plt.pie(market_share, labels=brands, explode=explode, autopct='%1.1f%%', startangle=90)

# Customize appearance
plt.title('Market Share of Brands')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Show plot
plt.show()
```

# Box Plot Styling

```python
import numpy as np

# Generate random data for the box plot
np.random.seed(0)
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Create box plot
plt.boxplot(data, 
            notch=True,  # Enable notched box plot
            vert=True,   # Plot vertically
            patch_artist=True,  # Fill boxes with color
            showfliers=True,    # Show outliers
            whis=1.5,  # Set whisker length
            boxprops=dict(facecolor='lightblue'),  # Set box fill color
            flierprops=dict(marker='o', color='red', markersize=8))  # Set outlier marker properties

# Customize appearance
plt.title('Box Plot with Custom Styling')
plt.xlabel('Data Sets')
plt.ylabel('Values')

# Show plot
plt.show()
```

# Multiple Subplots

```python
# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['Category A', 'Category B', 'Category C']
values = [10, 20, 15]
data = np.random.randn(100)

# Create figure with 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot line chart in top-left subplot
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Line Chart')

# Plot bar chart in top-right subplot
axs[0, 1].bar(categories, values)
axs[0, 1].set_title('Bar Chart')

# Plot histogram in bottom-left subplot
axs[1, 0].hist(data, bins=20, color='skyblue', edgecolor='black')
axs[1, 0].set_title('Histogram')

# Plot scatter plot in bottom-right subplot
axs[1, 1].scatter(x, y, color='red')
axs[1, 1].set_title('Scatter Plot')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show plot
plt.show()
```
